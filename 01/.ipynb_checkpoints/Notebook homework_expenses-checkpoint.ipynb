```python
from IPython.display import display, Markdown

expenses = [1200, 1500, 800, 600, 1000, 400, 900]

total = sum(expenses)
average = total / len(expenses)
maximum = max(expenses)
minimum = min(expenses)
count = sum(1 for x in expenses if x > 1000) 

display(Markdown(f"""
| Показник | Значення |
|---|---|
| Середнє значення | {average:.2f} |
| Максимальне значення | {maximum} |
| Мінімальне значення | {minimum} |
| Кількість (x > 1000) | {count} |
"""))

```
