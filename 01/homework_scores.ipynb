```python
data = {
    "Math": [80, 92, 75, 60, 88],
    "Physics": [70, 65, 90, 85, 78],
    "English": [95, 100, 85, 90, 87]
}

averages = {}

print(f"{'Предмет':<10} | {'Середній':<10} | {'Мін':<5} | {'Макс':<5}")
print("-" * 40)

for subject, scores in data.items():
    avg = sum(scores) / len(scores)
    min_score = min(scores)
    max_score = max(scores)
    
    averages[subject] = avg
    
    print(f"{subject:<10} | {avg:<10.2f} | {min_score:<5} | {max_score:<5}")
    
best_subject = max(averages, key=averages.get)

print("-" * 40)
```
