```python
from collections import Counter

# 1. Створення списку
votes = ["apple", "banana", "apple", "orange", "banana", "apple", "apple", "apple", "orange"]

# 2. Підрахунок голосів
counts = Counter(votes)

# 3. Визначення лідера
# most_common(1) повертає список із одним кортежем [('fruit', count)]
leader, leader_count = counts.most_common(1)[0]

# 4. Виведення результатів у таблиці
print(f"{'Фрукт':<10} | {'Кількість':<10}")
print("-" * 23)
for fruit, count in counts.items():
    print(f"{fruit:<10} | {count:<10}")

print("\n" + "="*23)
print(f"Лідер: {leader} ({leader_count} голосів)")
print("="*23 + "\n")

# 5. Топ-3 результати
top_3 = counts.most_common(3)
print("Топ-3 фрукти:")
for i, (fruit, count) in enumerate(top_3, 1):
    print(f"{i}. {fruit}: {count}")
```
